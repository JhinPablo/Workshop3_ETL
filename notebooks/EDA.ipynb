#Analisis, estandarización e imputación de columnas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar todos los archivos CSV desde la carpeta 'data'
df_2015 = pd.read_csv('C:/Users/admin-ilinf/Documents/ETL/Workshop3_ETL/data/2015.csv')
df_2016 = pd.read_csv('C:/Users/admin-ilinf/Documents/ETL/Workshop3_ETL/data/2016.csv')
df_2017 = pd.read_csv('C:/Users/admin-ilinf/Documents/ETL/Workshop3_ETL/data/2017.csv')
df_2018 = pd.read_csv('C:/Users/admin-ilinf/Documents/ETL/Workshop3_ETL/data/2018.csv')
df_2019 = pd.read_csv('C:/Users/admin-ilinf/Documents/ETL/Workshop3_ETL/data/2019.csv')


# Renombrar columnas para hacerlas consistentes
df_2015.rename(columns={
    'Happiness Rank': 'Happiness Rank',
    'Happiness Score': 'Score',
    'Economy (GDP per Capita)': 'GDP per capita',
    'Health (Life Expectancy)': 'Healthy life expectancy',
    'Trust (Government Corruption)': 'Perceptions of corruption',
    'Family': 'Social support'
}, inplace=True)

df_2016.rename(columns={
    'Happiness Rank': 'Happiness Rank',
    'Happiness Score': 'Score',
    'Economy (GDP per Capita)': 'GDP per capita',
    'Health (Life Expectancy)': 'Healthy life expectancy',
    'Trust (Government Corruption)': 'Perceptions of corruption',
    'Family': 'Social support'
}, inplace=True)

df_2017.rename(columns={
    'Happiness.Rank': 'Happiness Rank',
    'Happiness.Score': 'Score',
    'Economy..GDP.per.Capita.': 'GDP per capita',
    'Health..Life.Expectancy.': 'Healthy life expectancy',
    'Trust..Government.Corruption.': 'Perceptions of corruption',
    'Family': 'Social support'
}, inplace=True)

df_2018.rename(columns={
    'Overall rank': 'Happiness Rank',
    'Country or region': 'Country',
    'Score': 'Score',
    'GDP per capita': 'GDP per capita',
    'Social support': 'Social support',
    'Healthy life expectancy': 'Healthy life expectancy',
    'Freedom to make life choices': 'Freedom',
    'Generosity': 'Generosity',
    'Perceptions of corruption': 'Perceptions of corruption'
}, inplace=True)

df_2019.rename(columns={
    'Overall rank': 'Happiness Rank',
    'Country or region': 'Country',
    'Score': 'Score',
    'GDP per capita': 'GDP per capita',
    'Social support': 'Social support',
    'Healthy life expectancy': 'Healthy life expectancy',
    'Freedom to make life choices': 'Freedom',
    'Generosity': 'Generosity',
    'Perceptions of corruption': 'Perceptions of corruption'
}, inplace=True)

# Añadir columna de año a cada DataFrame
df_2015['Year'] = 2015
df_2016['Year'] = 2016
df_2017['Year'] = 2017
df_2018['Year'] = 2018
df_2019['Year'] = 2019

# Seleccionar columnas comunes para unificar
columns_to_keep = [
    'Country', 'Happiness Rank', 'Score', 'GDP per capita',
    'Social support', 'Healthy life expectancy', 'Freedom',
    'Generosity', 'Perceptions of corruption', 'Year'
]

# Asegurarse de que todas las columnas faltantes estén presentes en cada DataFrame
df_2015 = df_2015.reindex(columns=columns_to_keep)
df_2016 = df_2016.reindex(columns=columns_to_keep)
df_2017 = df_2017.reindex(columns=columns_to_keep)
df_2018 = df_2018.reindex(columns=columns_to_keep)
df_2019 = df_2019.reindex(columns=columns_to_keep)

# Combinar todos los DataFrames en uno solo
combined_df = pd.concat([df_2015, df_2016, df_2017, df_2018, df_2019], ignore_index=True)

# Imputar el valor nulo en 'Perceptions of corruption' con la media
combined_df['Perceptions of corruption'].fillna(combined_df['Perceptions of corruption'].mean(), inplace=True)

# Verificar si quedan valores nulos
print("Valores nulos por columna después de la imputación:")
print(combined_df.isnull().sum())

# Análisis de la importancia de las columnas mediante correlación con el puntaje de felicidad
numerical_columns = combined_df.select_dtypes(include=[np.number]).columns
correlation_matrix = combined_df[numerical_columns].corr()
happiness_correlation = correlation_matrix['Score'].sort_values(ascending=False)

# Visualizar la correlación de cada variable con el puntaje de felicidad
plt.figure(figsize=(10, 6))
sns.barplot(x=happiness_correlation.index, y=happiness_correlation.values)
plt.title('Correlación de las Variables con el Puntaje de Felicidad')
plt.xlabel('Columnas')
plt.ylabel('Correlación')
plt.xticks(rotation=45)
plt.show()

# Mostrar las correlaciones de las variables más importantes
print("Correlación con el Puntaje de Felicidad (Score):")
print(happiness_correlation)
```
